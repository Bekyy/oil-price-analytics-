```python
from statsmodels.tsa.stattools import adfuller

# Function to check stationarity using the Augmented Dickey-Fuller test
def is_stationary(data):
    result = adfuller(data)
    return result[1] <= 0.05  # p-value <= 0.05 means the data is stationary

# First-differencing each column
data_diff = combined_data_filled.diff().dropna()

# Check if differencing once is enough for stationarity
if not is_stationary(data_diff['Price']):  # Assuming you are checking stationarity on the 'Price' column
    data_diff = data_diff.diff().dropna()  # Second differencing if necessary
```

```python
# Now, try fitting the VAR model again with fewer lags and no high correlation

from statsmodels.tsa.api import VAR
from sklearn.preprocessing import StandardScaler

# Assuming we've reduced the variables based on correlation
data_diff_reduced = data_diff[['Price_diff', 'GDP']]  # Example: reduced to two variables

# Scale the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_diff_reduced)
data_diff_scaled = pd.DataFrame(data_scaled, columns=data_diff_reduced.columns, index=data_diff_reduced.index)

# Try fitting the VAR model with lower lags
model = VAR(data_diff_scaled)
results = model.fit(maxlags=3, ic='aic')  # Even fewer lags

# Forecasting for the next 10 time points
forecast = results.forecast(data_diff_scaled.values[-results.k_ar:], steps=10)
forecast_df = pd.DataFrame(forecast, index=pd.date_range(data.index[-1], periods=10, freq='M'), columns=data_diff_scaled.columns)

# Inverse transform to original scale if needed
forecast_original_scale = scaler.inverse_transform(forecast_df)
forecast_df = pd.DataFrame(forecast_original_scale, columns=data_diff_scaled.columns, index=forecast_df.index)

forecast_df
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import pandas as pd

# Step 1: Split data into training and testing sets
# Assuming we use the last 10 data points as the test set
test_size = 10
train_data = data_diff_scaled[:-test_size]
test_data = data_diff_scaled[-test_size:]

# Step 2: Fit the VAR model on the training data
model = VAR(train_data)
results = model.fit(maxlags=3, ic='aic')

# Step 3: Forecast for the length of the test set
lag_order = results.k_ar  # Get the optimal lag order determined by the model
forecast = results.forecast(train_data.values[-lag_order:], steps=test_size)
forecast_df = pd.DataFrame(forecast, index=test_data.index, columns=train_data.columns)

# Step 4: Inverse transform forecasts to original scale if needed
forecast_original_scale = scaler.inverse_transform(forecast_df)
forecast_df = pd.DataFrame(forecast_original_scale, columns=train_data.columns, index=forecast_df.index)

# Inverse transform the test data as well
test_data_original_scale = scaler.inverse_transform(test_data)
test_data_df = pd.DataFrame(test_data_original_scale, columns=train_data.columns, index=test_data.index)

# Step 5: Calculate evaluation metrics
mae = mean_absolute_error(test_data_df, forecast_df)
mse = mean_squared_error(test_data_df, forecast_df)
rmse = np.sqrt(mse)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")

# Optional: Plot the actual vs. forecasted values to visually inspect performance
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 3))
plt.plot(test_data_df['Price_diff'], label='Actual Price_diff', color='blue')
plt.plot(forecast_df['Price_diff'], label='Forecasted Price_diff', color='orange')
plt.legend()
plt.title('Actual vs Forecasted Brent Oil Price Difference')
plt.xlabel('Date')
plt.ylabel('Price Difference')
plt.show()
```

**Markov-Switching ARIMA Model**

```python
from statsmodels.tsa.regime_switching.markov_autoregression import MarkovAutoregression
import matplotlib.pyplot as plt

# Ensure data is stationary if needed and choose a column (e.g., 'Price_diff')
data_series = data_diff['Price_diff']

# Fit the Markov Autoregressive model with 2 regimes and autoregressive order 2
model = MarkovAutoregression(data_series, k_regimes=2, order=2)
result = model.fit()

# Display results summary
print(result.summary())

# Plot original data with smoothed regime probabilities
plt.figure(figsize=(12, 3))
plt.plot(data_series, label='Brent Oil Price Difference')
plt.plot(result.smoothed_marginal_probabilities[1], label='High-Volatility Regime Probability', color='orange')
plt.title('Brent Oil Prices with High-Volatility Regime Probability')
plt.xlabel('Date')
plt.ylabel('Price Difference')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from arch import arch_model
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot ACF and PACF for initial analysis
plt.figure(figsize=(12, 6))
plt.subplot(121)
plot_acf(data_diff['Price_diff'].dropna(), lags=20, ax=plt.gca())
plt.subplot(122)
plot_pacf(data_diff['Price_diff'].dropna(), lags=20, ax=plt.gca())
plt.show()

# Trying multiple ARIMA orders based on ACF/PACF
best_aic = np.inf
best_order = None
best_model = None

for p in range(3):  # trying up to p=2
    for d in range(2):  # differencing once or twice
        for q in range(3):  # trying up to q=2
            try:
                model = ARIMA(data_diff['Price_diff'], order=(p, d, q))
                results = model.fit()
                if results.aic < best_aic:
                    best_aic = results.aic
                    best_order = (p, d, q)
                    best_model = results
            except:
                continue

print(f"Best ARIMA order based on AIC: {best_order} with AIC={best_aic}")

# Fit the best ARIMA model and evaluate residuals
arima_model = ARIMA(data_diff['Price_diff'], order=best_order)
arima_result = arima_model.fit()
print(arima_result.summary())

# Fit GARCH model to ARIMA residuals
garch_model = arch_model(arima_result.resid.dropna(), vol='Garch', p=1, q=1)
garch_result = garch_model.fit()
print(garch_result.summary())

# Plot ARIMA trend and GARCH volatility
plt.figure(figsize=(12, 4))
plt.plot(data_diff.index, data_diff['Price_diff'], label='Brent Oil Price Difference')
plt.plot(data_diff.index, arima_result.fittedvalues, color='blue', label='ARIMA Trend')
plt.plot(data_diff.index, garch_result.conditional_volatility, color='green', label='GARCH Volatility')
plt.legend()
plt.title('ARIMA Trend and GARCH Volatility')
plt.show()
```

**LSTM Model for Time Series Forecasting**

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam

# Prepare data
data_scaled = data_diff[['Price_diff']].values  # Use only Price_diff for simplicity
#scaler = MinMaxScaler(feature_range=(0, 1))
#data_scaled = scaler.fit_transform(data_series)

# Create sequences
sequence_length = 12  # Use past 60 days to predict the next day
X, y = [], []
for i in range(len(data_scaled) - sequence_length):
    X.append(data_scaled[i:i + sequence_length])
    y.append(data_scaled[i + sequence_length])

X, y = np.array(X), np.array(y)

# Split into training and testing sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Build LSTM model
model = Sequential()
model.add(LSTM(units=64, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(LSTM(units=64))
model.add(Dense(1))

# Compile model
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

# Train model
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

# Forecast and evaluate
predicted = model.predict(X_test)
predicted = scaler.inverse_transform(predicted)
y_test_scaled = scaler.inverse_transform(y_test)

# Calculate RMSE
rmse = np.sqrt(np.mean((predicted - y_test_scaled) ** 2))
print(f'Root Mean Squared Error (RMSE): {rmse}')

# Plot actual vs predicted prices
plt.figure(figsize=(12,3))
plt.plot(y_test_scaled, label='Actual Price Difference', color='blue')
plt.plot(predicted, label='Predicted Price Difference', color='orange')
plt.title('Actual vs Predicted Price Differences')
plt.xlabel('Time')
plt.ylabel('Price Difference')
plt.legend()
plt.show()
```
